```python
from pydantic import BaseModel
from agentics.core.agentics import AG

class Answer(BaseModel):
    answer: str


class Question(BaseModel):
    question: str


async def main():
    agent = AG(atype=Answer, provide_explanations=True)
    response = await (
        agent
        << AG(
            atype=Question, states=[Question(question="What is the capital of France?")]
        )
    )
    for explanation in response.explanations:
        print(explanation.model_dump_json(indent=2))

await main()
```
